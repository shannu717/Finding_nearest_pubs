```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/open_pubs.csv',names=['fsa_id', 'name', 'address', 'postcode', 'easting', 'northing', 'latitude', 'longitude', 'local_authority'] )
```

```python
df
```

```python
df.isna().sum()
```

```python
# Replace "\n" with null values in the latitude and longitude columns
df.replace('\\N', np.nan, inplace=True)
```

```python
# Dropping rows which contain null values
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# Change datatype of latitude and Longitude to float 
df.latitude = df.latitude.astype(float)
df.longitude = df.longitude.astype(float)
df
```

```python
df.info()
```

```python
df.to_csv('pub.csv', index=False)
```
